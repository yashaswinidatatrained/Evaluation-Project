This dataset contain input features like
1.	R - This indicates Runs scored.
2.	AB - This means At bat or time at bat. 
3.	H - This means Hit. It's also called a "base hit".
4.	2B - This means the act of a batter striking the pitched ball and safely reaching second base without being called out by the umpire.
5.	3B - This means a Triple. It's is the act of a batter safely reaching third base after hitting the ball, with neither the benefit of a fielder's misplay nor another runner being put out on a fielder's choice.
6.	HR - This means Home runs. It's scored when the ball is hit in such a way that the batter is able to circle the bases and reach home plate safely in one play without any errors being committed by the defensive team.
7.	BB - This means Base on balls (also called a "walk").
8.	SO - Also denoted as "K" means Strikeout. 
9.	SB - This means Stolen base. 
10.	RA - This means Run Average. 
11.	ER - This means Earned run. 
12.	ERA - This means Earned Run Average. 
13.	CG - This means Complete Game.
14.	SHO - This means Shutout. 
15.	SV - This means Save. It's credited to a pitcher who finishes a game for the winning team under certain prescribed circumstances
16.	E - This means Errors. It's an act, in the judgment of the official scorer, of a fielder misplaying a ball in a manner that allows a batter or baserunner to advance one or more bases or allows a plate appearance to continue after the batter should have been put out.

The output target Variable: 
Number of predicted wins (W)

Since our target variable is continous data. This problem is considered as Regression problem and will build regression models.

We see there are outliers in R, ERA, SHO, Sv and E features, next we will treat the outliers.

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
np.where(z>3) #threshold = 3
z_df=df[(z<3).all(axis=1)]
print("After removing the outlier with z_score method the shape is: ", z_df.shape)
print("data_loss =   ", ((df.shape[0]-z_df.shape[0])/df.shape[0])*100)
```

We see the data loss is 3.3 and only one row is dropped in the Z_score method

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)

iqr = q3-q1

iqr_df = df[~((df <(q1 - 1.5*iqr))| (df>(q3+1.5*iqr))).any(axis=1)]
print("After removing the outlier with iqr method the shape is: ", iqr_df.shape)
print("data_loss= ", ((df.shape[0]-iqr_df.shape[0])/df.shape[0])*100)
```

We see that with iqr method data loss is 33.3 % which is very high, so we consider the data with the z_score method.

### Train_test_split

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
accuracy = 0
r_state =0
for i in range(1,500):
    X_train,X_test,y_train,y_test = train_test_split(x, y, test_size = 0.25, random_state = i)
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    y_pred = lr.predict(X_test)
    acc = r2_score(y_test,y_pred)
    if acc>accuracy:
        accuracy=acc
        r_state = i
print("max r2 score is ", accuracy , "randomstate = ",r_state)
```
